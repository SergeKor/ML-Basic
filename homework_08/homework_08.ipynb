```python
import numpy as np
import numpy.linalg as la
float_formatter = lambda x: "%.3f" % x
import matplotlib.pyplot as plt
import math

import PIL
%matplotlib inline
```

```python
n = 100
np.random.seed(42)  # фиксируем ПОРЯДОК случайностей
A = np.random.rand(n,n)

# SVD
U, S, VT = la.svd(A)

# сингулярные значения
#plt.plot(np.log(S))
```

```python
# Апроксимация
R = range(2,n)
err = np.zeros(n)
for r in R:
    sigma = np.diag(S)
    Aa = U[:, :r] @ sigma[:r, :r] @ VT[:r, :]
    er = math.sqrt(((A - Aa) ** 2).sum())
    err[r] = er
```

```python
plt.plot(R, err[2:])
plt.title = 'Зависимость ошибки апроксимации матрицы от ранга r'
```
