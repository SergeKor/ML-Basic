## Задание 1: Зеркальное отражение цифр

<br>

**Примеры:**
* `4  -> 0`

* `14 -> 4`

```python
days_to_vacation = 6

if days_to_vacation % 7 == 6:
    weekends = days_to_vacation // 7 * 2 + 1 
else:
    weekends = days_to_vacation // 7 * 2

weekends
```

## Задание 4: Римские числа

Пользователь вводит целое положительное число, программа должна вернуть строку в виде римского числа.

<br>

**Примеры:**
* `3   -> III`
* `15  -> XV`
* `234 -> CCXXXIV`

```python
# перевод в римское число до четырёхзначного как в примерах
number = 234

rim = ''

s = str(number)
l = len(s)
cif = int(s[-1])

if cif < 4:
    rim = 'I'*cif
else:
    if cif == 4 :
        rim = 'IV'
    else:    
        if cif == 5 :
            rim = 'V'
        else:
            if cif > 5 and cif < 9:
                rim = 'V' + 'I'*(cif-5)
            else:
                if cif == 9:
                    rim = 'IX'

s = s[:-1]
if s != '':
    cif = int(s[-1])

    if cif < 4:
        rim = 'X'*cif + rim
    else:
        if cif == 4 :
            rim = 'XL' + rim
        else:
            if cif == 5 :
                rim = 'L' + rim
            else:
                if cif > 4 and cif < 9:
                    rim = 'L' + 'X'*(cif-5) + rim
                else:
                    if cif == 9:
                        rim = 'XC' + rim

    s = s[:-1]
    if s != '':
        cif = int(s[-1])
        if cif < 4:
            rim = 'C'*cif + rim
        else:    
            if cif == 4 :
                rim = 'CD' + rim
            else:
                if cif == 5 :
                    rim = 'D' + rim
                else:
                    if cif > 5 and cif < 9:
                        rim = 'D' + 'C'*(cif-5) + rim
                    else:
                        if cif == 9:
                            rim = 'CM' + rim


rim
```

```python
def arabic_to_roman(number):
    thousands = ['', 'M', 'MM', 'MMM']
    hundreds = ['', 'C', 'CC', 'CCC', 'CD', 'D', 'DC', 'DCC', 'DCCC', 'CM']
    tens = ['', 'X', 'XX', 'XXX', 'XL', 'L', 'LX', 'LXX', 'LXXX', 'XC']
    units = ['', 'I', 'II', 'III', 'IV', 'V', 'VI', 'VII', 'VIII', 'IX']

    if not (1 <= number <= 3999):
        return "Недопустимое значение"

    t = thousands[number // 1000]
    h = hundreds[(number % 1000) // 100]
    te = tens[(number % 100) // 10]
    u = units[number % 10]

    #print(number // 1000 , (number % 1000) , (number % 100) , number % 10)
    return t + h + te + u

# Пример
print(arabic_to_roman(2234))  # CCXXXIV
```

## Задание 5: Проверка на вещественное число

Пользователь вводит данные, проверить - являются ли они положительным вещественным числом. Не использовать встроенные функции для проверки, только методы данных и конструкцию IF. (Дополнительное задание, по желанию - проверка на отрицательные вещественные числа)

<br>

**Примеры:**
* `5.6  -> True`
* `.78  -> True`
* `.67. -> False`
* `5    -> True`

```python
number = '.56.'

if len((number.split('.'))) < 3:
    print(True)
else:
    print(False)
```

```python
# Проверка на отрицательные вещественные числа
number = '-.56'

if len((number.split('.'))) < 3 and number[0]=='-':
    print(True)
else:
    print(False)
```
