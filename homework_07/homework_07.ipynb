### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
a77 = np.random.rand(7,7)
print('mean = ',a77.mean(),'std = ',a77.std())
min_val = np.min(a77)
max_val = np.max(a77)
normalized_arr = (a77 - min_val) / (max_val - min_val)
print('Нормализованный массив:',normalized_arr)
```

### Постройте две гистограммы распределения признака 'tip' в зависимости от категорий 'time'

```python
import matplotlib.pyplot as plt

kwargs = dict(histtype='stepfilled', alpha=0.3, density=True, bins=40)

tips_p = tips[['time','tip']]
for t in tips_p['time'].unique():    
    plt.hist(tips_p[tips_p['time']==t]['tip'], label=t, **kwargs)

plt.title("Распределение признака 'tip' в зависимости от категорий 'time'")
plt.legend()
plt.show()
```

### Постройте два графика scatterplot, представляющих взаимосвязь между признаками 'total_bill' и 'tip' один для Male, другой для Female и раскрасьте точки в зависимоти от признака 'smoker'

```python
fig, ax = plt.subplots(2)

tm = tips[tips['sex']=='Male'][['total_bill','tip','smoker']]
d = {0 : 'b', 1 :'r'}
tm['smoker_color']=tm['smoker'].map(d)

ax[0].scatter(tm['total_bill'], tm['tip'], c=tm['smoker_color'], cmap='viridis')
ax[0].set_title("Взаимосвязь между признаками 'total_bill' и 'tip' для Male и Female")
ax[0].set_ylabel('Male tip')


tm = tips[tips['sex']=='Female'][['total_bill','tip','smoker']]
d = {0 : 'b', 1 :'r'}
tm['smoker_color']=tm['smoker'].map(d)

ax[1].scatter(tm['total_bill'], tm['tip'], c=tm['smoker_color'], cmap='viridis', label = 'non smoker')
ax[1].set_ylabel('Female tip')

plt.xlabel('total_bill')
plt.legend()
```
